# Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

````python
import scipy.stats as stats
import numpy as np

# Given data
mean = 50
std_dev = 5
confidence_level = 0.95
sample_size = 30  # Assuming a sample size for the data

# Calculate the Z score for the 95% confidence interval
z_score = stats.norm.ppf((1 + confidence_level) / 2)

# Standard error
standard_error = std_dev / np.sqrt(sample_size)

# Calculate the confidence interval
margin_of_error = z_score * standard_error
confidence_interval = (mean - margin_of_error, mean + margin_of_error)

confidence_interval
````

# Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.

````python
from scipy.stats import chisquare

# Observed frequencies (data from the M&Ms bag)
# Replace these with actual data if available
observed = [22, 18, 20, 10, 12, 18]  # Blue, Orange, Green, Yellow, Red, Brown

# Expected frequencies based on the given percentages
total_mms = sum(observed)
expected = [0.20 * total_mms, 0.20 * total_mms, 0.20 * total_mms, 
            0.10 * total_mms, 0.10 * total_mms, 0.20 * total_mms]

# Perform the chi-square goodness of fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Significance level
alpha = 0.05

# Print the results
print(f"Chi-square statistic: {chi2_stat}")
print(f"p-value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis. The distribution of M&Ms does not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis. The distribution of M&Ms matches the expected distribution.")
````

# Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.

````python
import math
from scipy.stats import norm

# Given data
n = 500  # total population
smokers = 60  # number of individuals who smoke
confidence_level = 0.95

# Calculate the sample proportion
p_hat = smokers / n

# Z-score for 95% confidence level
z_score = norm.ppf((1 + confidence_level) / 2)

# Standard error of the proportion
standard_error = math.sqrt((p_hat * (1 - p_hat)) / n)

# Calculate the confidence interval
margin_of_error = z_score * standard_error
confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)

print(f"95% Confidence Interval: {confidence_interval}")
````

# Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.

````python
import numpy as np
from scipy.stats import norm

# Given data
mean = 75
std_dev = 12
confidence_level = 0.90
sample_size = 30  # Assuming a sample size of 30

# Z-score for 90% confidence interval
z_score = norm.ppf((1 + confidence_level) / 2)

# Standard error
standard_error = std_dev / np.sqrt(sample_size)

# Calculate the confidence interval
margin_of_error = z_score * standard_error
confidence_interval = (mean - margin_of_error, mean + margin_of_error)

print(f"90% Confidence Interval: {confidence_interval}")
````

This code will calculate the 90% confidence interval for the mean, showing a range where we are 90% confident that the true population mean lies within, based on the sample mean of 75.

# Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Degrees of freedom
df = 10

# Generate a range of x values (chi-square values)
x = np.linspace(0, 30, 1000)

# Compute the corresponding y values for the chi-square distribution
y = chi2.pdf(x, df)

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=f'Chi-square distribution (df={df})')

# Shade the area corresponding to chi-square statistic of 15
x_fill = np.linspace(15, 30, 500)
y_fill = chi2.pdf(x_fill, df)
plt.fill_between(x_fill, y_fill, color='red', alpha=0.5, label='Shaded area (statistic ≥ 15)')

# Add labels and title
plt.title('Chi-Square Distribution with 10 Degrees of Freedom')
plt.xlabel('Chi-Square Value')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
````

# Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.

To calculate the 99% confidence interval for the true proportion of people who prefer Coke, we use the formula for the confidence interval of a population proportion:

\[
CI = \hat{p} \pm Z \cdot \sqrt{\frac{\hat{p}(1 - \hat{p})}{n}}
\]

Where:
- \(\hat{p}\) is the sample proportion (520 out of 1000 people),
- \(Z\) is the Z-score for a 99% confidence level (approximately 2.576),
- \(n\) is the sample size (1000).

### Python Code:
Here's the code to calculate the 99% confidence interval for the true proportion of people who prefer Coke:

```python
import math
from scipy.stats import norm

# Given data
n = 1000  # total sample size
coke_preference = 520  # people who prefer Coke
confidence_level = 0.99

# Calculate the sample proportion
p_hat = coke_preference / n

# Z-score for 99% confidence level
z_score = norm.ppf((1 + confidence_level) / 2)

# Standard error of the proportion
standard_error = math.sqrt((p_hat * (1 - p_hat)) / n)

# Calculate the confidence interval
margin_of_error = z_score * standard_error
confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)

print(f"99% Confidence Interval: {confidence_interval}")
```

### Interpretation:
This code calculates the 99% confidence interval for the true proportion of people who prefer Coke. It will give you the range within which we are 99% confident that the true population proportion lies.

# Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.

To conduct a chi-square goodness of fit test for the given hypothesis, we need to compare the observed frequencies of the outcomes (tails and heads) with the expected frequencies under the assumption of a fair coin.

### Steps to Perform the Chi-Square Goodness of Fit Test

1. **Set up Hypotheses**:
   - Null Hypothesis (\(H_0\)): The coin is fair (the probability of heads = probability of tails = 0.5).
   - Alternative Hypothesis (\(H_a\)): The coin is biased (the probabilities are not equal).

2. **Observed Frequencies**:
   - Tails (observed) = 45
   - Heads (observed) = 100 - 45 = 55

   Therefore, the observed frequencies are:
   - Tails = 45
   - Heads = 55

3. **Expected Frequencies** (under the null hypothesis of a fair coin):
   - Tails (expected) = \(100 \times 0.5 = 50\)
   - Heads (expected) = \(100 \times 0.5 = 50\)

4. **Chi-Square Test Statistic**:
   The formula for the chi-square statistic is:

   \[
   \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
   \]

   where \(O_i\) is the observed frequency, and \(E_i\) is the expected frequency.

5. **Degrees of Freedom**:
   Degrees of freedom for this test is given by:
   \[
   df = k - 1
   \]
   where \(k\) is the number of categories (in this case, 2: heads and tails). So, \(df = 2 - 1 = 1\).

6. **Critical Value and Conclusion**:
   We compare the calculated chi-square statistic to the critical value from the chi-square distribution table at \(df = 1\) and significance level \(\alpha = 0.05\).

### Calculations

Now let's calculate the chi-square statistic. 

\[
\chi^2 = \frac{(45 - 50)^2}{50} + \frac{(55 - 50)^2}{50}
\]

\[
\chi^2 = \frac{(-5)^2}{50} + \frac{(5)^2}{50}
\]

\[
\chi^2 = \frac{25}{50} + \frac{25}{50} = 0.5 + 0.5 = 1
\]

Now we will check the critical value and make our conclusion. 

### Critical Value

Using a chi-square distribution table, the critical value for \(df = 1\) at \(\alpha = 0.05\) is approximately **3.841**.

### Conclusion

- **Calculated Chi-Square Statistic**: \(1\)
- **Critical Value**: \(3.841\)

Since \(1 < 3.841\), we fail to reject the null hypothesis. 

### Result

There is not enough evidence to conclude that the coin is biased towards tails at the 0.05 significance level. The observed frequencies do not significantly differ from what we would expect if the coin were fair.

# Q9. A study was conducted to determine if there is an association between smoking status (smoker or non-smoker) and lung cancer diagnosis (yes or no). The results are shown in the contingency table below. Conduct a chi-square test for independence to determine if there is a significant association between smoking status and lung cancer diagnosis.

Use a significance level of 0.05.
Group A

Outcome 1 20 15
Outcome 2 10 25
Outcome 3 15 20
Group B

Lung Cancer: Yes

Smoker 60 140
Non-smoker 30 170

To determine if there is a significant association between smoking status and lung cancer diagnosis using a chi-square test for independence, we will follow these steps:

### 1. Set up Hypotheses

- **Null Hypothesis (\(H_0\))**: There is no association between smoking status and lung cancer diagnosis (they are independent).
- **Alternative Hypothesis (\(H_a\)**: There is an association between smoking status and lung cancer diagnosis (they are not independent).

### 2. Organize the Data

From the given data, we can summarize it in a contingency table:

|               | Lung Cancer: Yes | Lung Cancer: No | Row Total |
|---------------|------------------|------------------|-----------|
| **Smoker**    | 60               | 140              | 200       |
| **Non-smoker**| 30               | 170              | 200       |
| **Column Total** | 90            | 310              | 400       |

### 3. Calculate Expected Frequencies

The expected frequency for each cell can be calculated using the formula:

\[
E_{ij} = \frac{(Row \ total_i) \times (Column \ total_j)}{Grand \ total}
\]

- For Smoker, Lung Cancer: Yes:
  \[
  E_{11} = \frac{200 \times 90}{400} = 45
  \]
  
- For Smoker, Lung Cancer: No:
  \[
  E_{12} = \frac{200 \times 310}{400} = 155
  \]
  
- For Non-smoker, Lung Cancer: Yes:
  \[
  E_{21} = \frac{200 \times 90}{400} = 45
  \]

- For Non-smoker, Lung Cancer: No:
  \[
  E_{22} = \frac{200 \times 310}{400} = 155
  \]

### 4. Complete Contingency Table

Now, we have:

|               | Lung Cancer: Yes | Lung Cancer: No | Row Total |
|---------------|------------------|------------------|-----------|
| **Smoker**    | 60 (O)           | 140 (O)          | 200       |
|               | 45 (E)           | 155 (E)          |           |
| **Non-smoker**| 30 (O)           | 170 (O)          | 200       |
|               | 45 (E)           | 155 (E)          |           |
| **Column Total** | 90            | 310              | 400       |

### 5. Calculate Chi-Square Statistic

The chi-square statistic is calculated using the formula:

\[
\chi^2 = \sum \frac{(O_{ij} - E_{ij})^2}{E_{ij}}
\]

Calculating for each cell:

- For Smoker, Lung Cancer: Yes:
  \[
  \frac{(60 - 45)^2}{45} = \frac{15^2}{45} = \frac{225}{45} = 5
  \]

- For Smoker, Lung Cancer: No:
  \[
  \frac{(140 - 155)^2}{155} = \frac{(-15)^2}{155} = \frac{225}{155} \approx 1.45
  \]

- For Non-smoker, Lung Cancer: Yes:
  \[
  \frac{(30 - 45)^2}{45} = \frac{(-15)^2}{45} = \frac{225}{45} = 5
  \]

- For Non-smoker, Lung Cancer: No:
  \[
  \frac{(170 - 155)^2}{155} = \frac{15^2}{155} = \frac{225}{155} \approx 1.45
  \]

Now, sum these values:

\[
\chi^2 = 5 + 1.45 + 5 + 1.45 \approx 13.9
\]

### 6. Degrees of Freedom

The degrees of freedom for a chi-square test for independence is given by:

\[
df = (r - 1)(c - 1)
\]

where \(r\) is the number of rows and \(c\) is the number of columns. Here, we have \(r = 2\) (smoker, non-smoker) and \(c = 2\) (lung cancer: yes, no):

\[
df = (2 - 1)(2 - 1) = 1
\]

### 7. Critical Value and Conclusion

Using a chi-square distribution table, the critical value for \(df = 1\) at \(\alpha = 0.05\) is approximately **3.841**.

### Conclusion

- **Calculated Chi-Square Statistic**: \( \approx 13.9 \)
- **Critical Value**: \(3.841\)

Since \(13.9 > 3.841\), we reject the null hypothesis.

### Result

There is a significant association between smoking status and lung cancer diagnosis at the 0.05 significance level.

# Q10. - **Null Hypothesis (\(H_0\))**: There is no association between chocolate preference and country of origin (they are independent).
- **Alternative Hypothesis (\(H_a\))**: There is an association between chocolate preference and country of origin (they are not independent).

### 3. Calculate Expected Frequencies

\[
E_{ij} = \frac{(Row \ total_i) \times (Column \ total_j)}{Grand \ total}

  \[

### 4. Complete Contingency Table

| Chocolate Preference | U.S. (O) | U.K. (O) | U.S. (E) | U.K. (E) |
|----------------------|----------|----------|----------|----------|
| Milk Chocolate       | 200      | 225      | 212.5    | 212.5    |
| Dark Chocolate       | 150      | 175      | 162.5    | 162.5    |
| White Chocolate      | 150      | 100      | 125      | 125      |

### 5. Calculate Chi-Square Statistic

The chi-square statistic is calculated using the formula:

\[
\chi^2 = \sum \frac{(O_{ij} - E_{ij})^2}{E_{ij}}
\]

Calculating for each cell:

- For Milk Chocolate (U.S.):
  \[
  \frac{(200 - 212.5)^2}{212.5} = \frac{(-12.5)^2}{212.5} = \frac{156.25}{212.5} \approx 0.735
  \]

- For Milk Chocolate (U.K.):
  \[
  \frac{(225 - 212.5)^2}{212.5} = \frac{(12.5)^2}{212.5} = \frac{156.25}{212.5} \approx 0.735
  \]

- For Dark Chocolate (U.S.):
  \[
  \frac{(150 - 162.5)^2}{162.5} = \frac{(-12.5)^2}{162.5} = \frac{156.25}{162.5} \approx 0.962
  \]

- For Dark Chocolate (U.K.):
  \[
  \frac{(175 - 162.5)^2}{162.5} = \frac{(12.5)^2}{162.5} = \frac{156.25}{162.5} \approx 0.962
  \]

- For White Chocolate (U.S.):
  \[

Now, summing these values:

\[

\]

### 6. Degrees of Freedom

The degrees of freedom for a chi-square test for independence is given by:

\[

where \(r\) is the number of rows and \(c\) is the number of columns. Here, we have \(r = 3\) (chocolate preferences) and \(c = 2\) (U.S. and U.K.):

\[

# Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.

To conduct a hypothesis test to determine if the population mean is significantly different from 70 based on the sample data, we can use a one-sample t-test. Here are the steps involved:

### 1. Set Up Hypotheses

- **Null Hypothesis (\(H_0\))**: The population mean is equal to 70 (\(\mu = 70\)).
- **Alternative Hypothesis (\(H_a\))**: The population mean is not equal to 70 (\(\mu \neq 70\)).

### 2. Collect Sample Data

Given data:
- Sample size (\(n\)) = 30
- Sample mean (\(\bar{x}\)) = 72
- Sample standard deviation (\(s\)) = 10

### 3. Calculate the Test Statistic

We use the formula for the t-statistic:

\[
t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}
\]

where:
- \(\bar{x}\) = sample mean
- \(\mu_0\) = hypothesized population mean (70)
- \(s\) = sample standard deviation
- \(n\) = sample size

Substituting the values:

\[
t = \frac{72 - 70}{10 / \sqrt{30}} = \frac{2}{10 / 5.477} = \frac{2}{1.825} \approx 1.095
\]

### 4. Determine Degrees of Freedom

The degrees of freedom (\(df\)) for a one-sample t-test is given by:

\[
df = n - 1 = 30 - 1 = 29
\]

### 5. Find the Critical t-value

Using a t-distribution table, we find the critical t-value for a two-tailed test at a significance level of 0.05 with 29 degrees of freedom. The critical t-value is approximately:

\[
t_{critical} \approx \pm 2.045
\]

### 6. Compare the Test Statistic to the Critical Value

- **Calculated t-statistic**: \( \approx 1.095 \)
- **Critical t-value**: \( \pm 2.045 \)

### 7. Conclusion

Since the calculated t-statistic (\(1.095\)) does not exceed the critical t-value (\(2.045\)), we fail to reject the null hypothesis.

### Final Result

At the 0.05 significance level, there is not enough evidence to conclude that the population mean is significantly different from 70.
